```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

# Importing Data

```python
df=pd.read_csv('USA_Housing.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# Checking Null Value

```python
df.isna().sum()
```

# divide dataset into x & y

```python
x=df.iloc[:,:5].values
x
```

```python
y=df.iloc[:,5].values
y
```

# divide into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2, random_state = 101)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

#  Creating Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg=LinearRegression()
linreg.fit(xtrain,ytrain)
```

#  Model Evaluation

 interpretation of intercepts

```python
linreg.intercept_
```

Regression cofficients

```python
linreg.coef_
```

```python
coef_df = pd.DataFrame(linreg.coef_ , x.columns , columns=['Coefficient'])
coef_df
```

# Interpreting the coefficient of regression 

```python
seting all other x constant, if we increase Avg. Area Income by 1 unit, the y will increase by 21.566696
seting all other x constant, if we increase Avg. Area House Age by 1 unit, the y will increase by 165453.042478 
seting all other x constant, if we increase Avg. Area Number of Rooms by 1 unit, the y will increase by 120499.839093 
seting all other x constant, if we increase Avg. Area Number of Bedrooms by 1unit, the y will increase by 1999.785336
seting all other x constant, if we increase Area Population by 1unit, the y will increase by 15.340604
```

#  Predictions for our model

```python
ypred = linreg.predict(xtest)
ypred
```

```python
plt.scatter(ytest,ypred)
```

#  Regression Evaluation Metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print(f'MAError-:{mean_absolute_error(ytest,ypred)}')
print(f'MSEError-:{mean_squared_error(ytest,ypred)}')
print(f'RMAError-:{np.sqrt(mean_squared_error(ytest,ypred))}')
print(f'r2-:{r2_score(ytest,ypred)}')
```
